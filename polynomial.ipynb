```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('C:\\Users\\omkar\\Downloads\\data.csv')
x=np.array(data.Temperature).reshape(-1,1)
y=np.array(data.Pressure).reshape(-1,1)
data=data.dropna()
```

```python
degree=3
polynomial_features=PolynomialFeatures(degree=degree)
x_new=polynomial_features.fit_transform(x)
alpha=1.0
ridge_model=Ridge(alpha=alpha)
ridge_model.fit(x_new,y)
Y_new=ridge_model.predict(x_new)
plt.scatter(x,y,color='red')
plt.plot(x,Y_new)
plt.show()
mse=mean_squared_error(y,Y_new)
print("mean squared error-",mse)
mae=mean_absolute_error(y,Y_new)
print("mean absolute error-",mae)
```
