```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
data=pd.read_csv('C:\\Users\\omkar\\Downloads\\student_info.csv')
data=data.dropna()
x=np.array(data.study_hours[0:100]).reshape(100,1)
y=np.array(data.student_marks[0:100]).reshape(100,1)
model=LinearRegression()
model.fit(x,y)
y_pred=model.predict(x)
plt.scatter(x,y,color='green')
plt.plot(x,y_pred,color='red')
plt.xlabel('Hours of Study')
plt.ylabel('Marks')
plt.legend()
plt.show()
```
