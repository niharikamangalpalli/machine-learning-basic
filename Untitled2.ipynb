```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
data=pd.read_csv('C:\\Users\\omkar\\Downloads\\data.csv')
x=np.array(data.Temperature[0:6]).reshape(6,1)
y=np.array(data.Pressure[0:6]).reshape(6,1)
degree=3
polynomial_features=PolynomialFeatures(degree=degree)
x_poly=polynomial_features.fit_transform(x)
model=LinearRegression()
model.fit(x_poly,y)
y_poly_pred=model.predict(x_poly)
plt.plot(x,y_poly_pred,color='r')
plt.title(f'polynomial Regression (degree={degree})')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.show()
mse=mean_squared_error(y,y_poly_pred)
print(f'Mean squared Error:{mse}')
inp=int(input("Enter the temperature"))
inp_poly=polynomial_features.fit_transform(np.array([[inp]]))
print(model.predict(inp_poly))
```
